```python
#no of employees promoted
res=train['is_promoted'].value_counts()
total=train['is_promoted'].count()
notpromoted=res[0]
promoted=res[1]
print(notpromoted,promoted)
```

```python
#findings the % of employees promoted
proper=(promoted/total)*100
print("Percentage of promoted employees is:{:.2f}%".format(proper))
```

```python
#plot to show gap in promoted and non-promoted employees
plt.hist(train['is_promoted'])
plt.title("plot to show gap in promted and non-promoted employees",fontsize=30)
plt.xlabel('0-non promoted and 1 promotes',fontsize=20)
plt.grid()
plt.ylabel('count')
plt.show()
```

```python
#dist of avg training score
plt.rcParams['figure.figsize']=(15,7)
sns.histplot(train['avg_training_score'],color='cyan')
plt.title('dist of avg training score',fontsize=30)
plt.xlabel('Avg training score')
plt.ylabel('count')
plt.show()
```

```python
#counting the number of employees who have won the award
res=train['awards_won?'].value_counts()
size=[res[0],res[1]]
exp=[0,.2]
color=['red','green']
labels=['NO Award won','Award Won']
plt.pie(size,colors=color,autopct='%0.2f%%',labels=labels,radius=.8,explode=exp)
plt.title("Showing % of employees who won award",fontsize=20)
plt.show()
```

```python
#checking the distribution of length of service
sns.distplot(train['length_of_service'],color='navy')
plt.title('checking the distribution of length of service',fontsize=30)
plt.xlabel("Length of service in years",fontsize=15)
plt.ylabel('count')
plt.show()
```

```python
#checking the distripbution of age of employees of company 

sns.distplot(train['age'],color='pink')
plt.title('checking the distribution of age of employees',fontsize=30)
plt.ylabel('count')
plt.show()
```

```python
# checking the different type of recuritment channel for the company
res=train['recruitment_channel'].value_counts()
size=[res['other'],res['sourcing'],res['referred']]
color=["red","green","blue"]
labels=["others","sourcing","reffered"]


plt.rcParams['figure.figsize']=(9,9)
plt.pie(size,colors=color,labels=labels,shadow=True,autopct="%0.2f%%")
plt.title("Showing share of different recuritment channels",fontsize=30)
my_circle=plt.Circle((0,0),0.7,color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.legend()
plt.show()
# print(size)
```
